```python
'''
From the generated Info, our datatypes are not due
So we will first convert the different columns to their ideal datatypes
'''

for i in df_transaction.columns:
    if i in ['transaction_id', 'product_id', 'customer_id', 'list_price', 'standard_cost', 'product_first_sold_date']:
        df_transaction[i] = pd.to_numeric(df_transaction[i])
        
    elif i in ['transaction_date']:
        df_transaction[i] = pd.to_datetime(df_transaction[i], format='%Y-%m-%d')
        
df_transaction.info()
```

```python
for i in df_new_customer.columns:
    if i in ['past_3_years_bike_related_purchases', 'tenure','postcode', 'Rank', 'Value']:
        df_new_customer[i] = pd.to_numeric(df_new_customer[i])
        
    elif i in ['DOB']:
        df_new_customer[i] = pd.to_datetime(df_new_customer[i])
        
    
df_new_customer.info()
```
